# GET DATASET

```python
col_name=['user_id','item_id','rating','timestamp']
df=pd.read_csv('ml-100k/u.data' ,sep='\t',names=col_name)
df.head()
```

```python
df.shape
```

```python
df['user_id'].nunique()
```

```python
df['item_id'].nunique()
```

```python
movie_titles=pd.read_csv('ml-100k/u.item' ,sep='\|',header=None)
movie_titles.head()
```

```python
movie_titles.shape
```

```python
movie_title=movie_titles[[0,1]]
movie_title.columns=["item_id",'title']
```

```python
movie_title.head()
```

```python
df=pd.merge(df,movie_title,on='item_id')
```

```python
df.tail()
```

## Making Recommendation System

```python
movie_matrix=df.pivot_table(index='user_id',columns='title' ,values='rating')
movie_matrix
```

```python
starwar_user_rating=movie_matrix['Star Wars (1977)']
starwar_user_rating.head()
```

```python
similar_to_starwar=movie_matrix.corrwith(starwar_user_rating)
```

```python
corr_starwar=pd.DataFrame(similar_to_starwar,columns=['correlation'])
```

```python
corr_starwar
```

```python
corr_starwar.dropna(inplace=True)
```

```python
corr_starwar
```

```python
corr_starwar=corr_starwar.join(rating['no of rating'])
```

```python
corr_starwar.head()
```

```python
corr_starwar[corr_starwar['no of rating']>100].sort_values('correlation' ,ascending=False)
```
