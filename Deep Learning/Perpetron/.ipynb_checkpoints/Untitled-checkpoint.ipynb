```python
#x,y=make_blobs(n_samples=500,centers=2,n_features=2,random_state=11)
x,y=make_moons(n_samples=1000,shuffle=True, noise=0.2, random_state=1)
print(x.shape,y.shape)
```

```python
plt.style.use("seaborn")
plt.scatter(x[:,0],x[:,1],c=y,cmap=plt.cm.Accent)
plt.show()
```

### define model and helper functions

```python
def sigmoid(z):
    return (1.0)/(1+np.exp(-z))
```

```python
#broadcasting 
z=np.array([1,2,3,4])
sigmoid(z)

a=np.array([[2,3,4,5],
            [10,20,30,40]])
print(np.dot(a,z))
print((a*z))
```

```python
def predict(x,weights):
    z=np.dot(x,weights)
    preds= sigmoid(z)
    return preds

def loss(x,y,weights):
    y_=predict(x,weights)
    cost=np.mean(-y*np.log(y_)-(1-y_)*np.log(1-y_))
    return cost

def update(x,y,weights,learning_rate): 
    
    #we calculate gradient here 
    
    y_=predict(x,weights)
    delta_w=np.dot(x.T,y_-y)
    m=x.shape[0]
    
    weights=weights-learning_rate*delta_w/(float(m))
    return weights
def train(x,y,learning_rate=0.5,max_epochs=100):
    
    #modify input matrix by adusting bias 
    ones=np.ones((x.shape[0],1))
    x=np.hstack((ones,x))
    
    weights=np.zeros(x.shape[1])
    
    for i in range(max_epochs):
        weights=update(x,y,weights,learning_rate)
        
        #print progress after every 10 epochs 
        
        if i%10==0:
            l=loss(x,y,weights)
            print("Ecpoch %d loss %.4f"%(i,l))
    return weights
    
    
    
    
    
```

```python
weights=train(x,y,learning_rate=0.8,max_epochs=2000)
```

```python
def getprediction(x_test,weights,labels=True):
    if x_test.shape[1]!=weights.shape[0]:
        ones=np.ones((x_test.shape[0],1))
        x_test=np.hstack((ones,x_test))
    preds=predict(x_test,weights)
    
    if not labels:
        return preds
    else:
        labels=np.zeros(preds.shape)
        labels[preds>=0.5]=1
        return labels
        
        
    
```

```python
x1=np.linspace(-2,2,10)
x2=-(weights[0]+weights[1]*x1)/weights[2]
```
